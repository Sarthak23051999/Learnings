```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def get_vif_factors(X):
    vif_factors = pd.DataFrame()
    X_matrix = X.values
    vif  = [variance_inflation_factor(X_matrix,i) for i in range(X_matrix.shape[1])]
    vif_factors['column'] = X.columns
    vif_factors['vif'] = vif
    vif_factors = vif_factors.sort_values('vif',ascending=True)
    return vif_factors
```

```python
for x in ipl_df_new.columns:
    print(ipl_df_new[x].dtype)
```

```python
vif_factors = get_vif_factors(ipl_df_new)
```

```python
plt.figure(figsize=(15,13))
sn.heatmap(ipl_df_new[vif_factors[vif_factors['vif']>4]['column']].corr(),annot=True)
plt.title("Heatmap depicting co-relation between features")

#CAPTAINCY_EXP, T_RUNS
#HS, AVE, RUN_S, SIXERS, ROLE_BOWLER, SR-B
#AGE_3, AGE_2
#T_WKTS,ODI_WKTS
#WKTS, RUN-C
#AVE-BL, SR-BL,ECON
```

```python
## Takes model as parameter
# print the rmse of model
from sklearn import metrics
def get_train_test_rmse(model):
    y_train_pred = model.predict(X_train)
    rmse_train = round(np.sqrt(metrics.mean_squared_error(y_train,y_train_pred)),3)
    y_test_pred = model.predict(X_test)
    rmse_test = round(np.sqrt(metrics.mean_squared_error(y_test,y_test_pred)),3)
    print("Train rmse : ",rmse_train)
    print("Test rmse : ",rmse_test)
```

```python
get_train_test_rmse(linreg)
```

## Regularization

```python
#alpha indiacates strength of  regularization
from sklearn.linear_model import Ridge
ridge_alpha1 = Ridge(alpha = 1,  max_iter = 500)
ridge_alpha1.fit(X_train, y_train)
ridge_alpha2 = Ridge(alpha = 2,  max_iter = 500)
ridge_alpha2.fit(X_train, y_train)
```

```python
print(get_train_test_rmse(ridge_alpha1))
print(get_train_test_rmse(ridge_alpha2))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha = 0.01,max_iter=500)
lasso.fit(X_train,y_train)
print(get_train_test_rmse(lasso))
```
